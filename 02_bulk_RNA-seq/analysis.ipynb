```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("synthetic_expression.csv", index_col=0)
df

```

```python
df.info()
```

```python
X = df.values         
X
gene_names = df.index

```

```python
pca = PCA(n_components=2)
pc_values = pca.fit_transform(X)
pc_values



```

```python
plt.scatter(pc_values[:, 0], pc_values[:, 1])

for i, gene in enumerate(gene_names):
    plt.text(pc_values[i, 0], pc_values[i, 1], gene)

plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0] * 100:.2f}% variance)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1] * 100:.2f}% variance)")
plt.title("PCA of Larger Synthetic Gene Expression")
plt.show()
```
